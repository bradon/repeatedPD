#Imports, Standard Function

```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
%pylab inline
file_path_prefix="E:\\DataSet1\\Grid"

def standardTimeSeries(set_number, delta, r):
    filename="_set_"+str(set_number)+"_delta_"+str(delta)+"_r_"+str(r)+".csv"
    filename=file_path_prefix+filename
    df=pd.read_csv(filename,index_col=0,error_bad_lines=False)
    #df["totalPayoff"]=df["totalPayoff"]/250)
    df.plot(figsize=(10,3))
    plt.title("Delta="+str(delta)+", r="+str(r))
    plt.ylabel("Payoff")
    plt.legend().set_visible(False)
    plt.xlabel("Generation")
    plt.ylim((240.0, 510.0))
```
